# Pir chart for  Return and replacement policy of the e-tailer is important for purchase decision for online shopping with the count

    # Most of the users are ##Strongly Agree for Return and replacement policy of the e-tailer is important of purchase decision

```python
lst = []
label = []
cont = []
explode = (0,0,0.1)
data.groupby(columns[32]).groups.keys()
for i in data.groupby(columns[32]).groups.keys():
    dd = data[(data[columns[32]] == i )]
    lst.append([i,dd[columns[32]].count()])
for i in lst:
    label.append(i[0])
    cont.append(i[1])

# Creating plot
fig = plt.figure(figsize =(10, 7))
plt.pie(cont,explode=explode,labels = label,autopct='%1.1f%%',shadow=True)
plt.title('Count of Customers who thinks Return and replacement policy of the e-tailer is important for purchase decision for online shopping  ')
# show plot
plt.show()
```
