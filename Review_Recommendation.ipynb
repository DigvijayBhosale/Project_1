# Sting Cleaning

```python
import nltk
from nltk.stem import PorterStemmer,WordNetLemmatizer

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
nltk.download('wordnet')

punctuation_new=string.punctuation.replace('\'','')+'0123456789'
rm_punct = str.maketrans(punctuation_new, ' ' * len(punctuation_new))
```

```python
review_main = []
for char in dataframe_review.REVIEW:
    review_main.append(char)
for i in range(len(review_main)):
    p=[]
    for word in review_main[i].split():
        word=word.lower()
        if word in words_dict:
            p.append(words_dict[word].lower())
        else:
            p.append(word)
        review_main[i]   = ' '.join(p)
for i in range(len(review_main)):
    review_main[i] = review_main[i].replace('\n',' ').replace('\t',' ')
    review_main[i] = re.sub(r'http[s]?://(?:[a-z]|[0-9]|[$-_@.&+~]|[!*\(\),~]|(~?:%[0-9a-f][0-9a-f]))+', '', review_main[i])
    review_main[i] = review_main[i].translate(rm_punct)
    split_words=[]
    for word in review_main[i].split():
        if word.lower() not in req_stopwords:
            split_words.append(stemmer.stem(lemmatizer.lemmatize(word,pos='v')))
    review_main[i]   = ' '.join(split_words)

print('Special Char and Digits Removed from comments : ',punctuation_new)    
print('Removed stopwords comments : ',punctuation_new)
```
